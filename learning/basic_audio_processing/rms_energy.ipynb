```python
import numpy as np
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load audio files
debussy_file = "./audio/debussy.wav"
redhot_file = "./audio/redhot.wav"
duke_file = "./audio/duke.wav"

debussy, sr = librosa.load(debussy_file)
redhot, sr = librosa.load(redhot_file)
duke, sr = librosa.load(duke_file)
```

```python
# extract RMSE with librosa

FRAME_LENGTH = 1024
HOP_LENGTH = 512

rms_debussy = librosa.feature.rms(debussy, frame_length=FRAME_LENGTH, hop_length=HOP_LENGTH)[0]
rms_redhot = librosa.feature.rms(redhot, frame_length=FRAME_LENGTH, hop_length=HOP_LENGTH)[0]
rms_duke = librosa.feature.rms(duke, frame_length=FRAME_LENGTH, hop_length=HOP_LENGTH)[0]
```

```python
# Plot the RMSE  for all audio files

# Visualize amplitude envelope for all the audio files

frames = range(0, rms_debussy.size)
t = librosa.frames_to_time(frames, hop_length=HOP_LENGTH)

plt.figure(figsize=(15, 17))

plt.subplot(3, 1, 1)
librosa.display.waveshow(debussy, alpha=0.5)
plt.plot(t, rms_debussy, color='r')
plt.title("Debussy")
plt.ylim(-1, 1)

plt.subplot(3, 1, 2)
librosa.display.waveshow(redhot, alpha=0.5)
plt.plot(t, rms_redhot, color='r')
plt.title("Debussy")
plt.ylim(-1, 1)

plt.subplot(3, 1, 3)
librosa.display.waveshow(duke, alpha=0.5)
plt.plot(t, rms_duke, color='r')
plt.title("Debussy")
plt.ylim(-1, 1)
```

```python
def rms_energy(signal, frame_length, hop_length):
    rms = []

    for i in range(0, len(signal), hop_length):
        rms_current_frame = np.sqrt(np.sum(signal[i:i+frame_length] ** 2 / frame_length))
        rms.append(rms_current_frame)
    return np.array(rms)
```

```python
rms_debussy = rms_energy(debussy, FRAME_LENGTH, HOP_LENGTH)
rms_redhot = rms_energy(redhot, FRAME_LENGTH, HOP_LENGTH)
rms_duke = rms_energy(duke, FRAME_LENGTH, HOP_LENGTH)
```

```python
# Plot the RMSE  for all audio files

# Visualize amplitude envelope for all the audio files

frames = range(0, rms_debussy.size)
t = librosa.frames_to_time(frames, hop_length=HOP_LENGTH)

plt.figure(figsize=(15, 17))

plt.subplot(3, 1, 1)
librosa.display.waveshow(debussy, alpha=0.5)
plt.plot(t, rms_debussy, color='r')
plt.title("Debussy")
plt.ylim(-1, 1)

plt.subplot(3, 1, 2)
librosa.display.waveshow(redhot, alpha=0.5)
plt.plot(t, rms_redhot, color='r')
plt.title("Debussy")
plt.ylim(-1, 1)

plt.subplot(3, 1, 3)
librosa.display.waveshow(duke, alpha=0.5)
plt.plot(t, rms_duke, color='r')
plt.title("Debussy")
plt.ylim(-1, 1)
```
